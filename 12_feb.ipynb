```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]
x2 = np.arange(0, 4.5, 0.5)
plt.plot(x, y, 'b^--', label = '2*x')
plt.plot(x2[:6], x2[:6]**2, 'r-', label = 'x^2')
plt.plot(x2[5:], x2[5:]**2, 'r--')
plt.title("Our first graph!")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
```
